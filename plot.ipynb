```python
def get_best_spec(spec='cover3'):
    """To get the best spec models for all environments"""
    log_keys = ['DubinsCar_postignore', 'SingleIntegrator_postignore', 'DoubleIntegrator_postignore']
    all_dfs = [pd.read_csv(csv_file_2load) for log_key in log_keys for csv_file_2load in logs[log_key] ]
    df = pd.concat(all_dfs, ignore_index=True)
    df = df.sort_values('success_mean',ascending=False)  #  pick best one in each model
    df = df.dropna()
    
#     df = df[df['num_agents'] <=8]
    
    
    df['env'] = df["path"].apply(lambda x: x.split('/')[6])
    df["path"] = df["path"].apply(lambda x: '/'.join(x.split('/')[4:]))
    
    df = df[df['spec'] == spec]
    df_pivoted = df.pivot_table(index=['env','num_agents'], columns='planner', values= ['path'],
                               aggfunc='first')
#     df_pivoted = df_grouped.first().unstack(level='planner')
    df_pivoted.columns = df_pivoted.columns.swaplevel(0, 1)
    df_pivoted = df_pivoted.sort_index(axis=1, level=0)
    
    return df_pivoted
```

```python
spec_map = (('seq', 'Sequence'), ('cover', 'Cover'),('2loop', 'Loop'), ('2branch', 'Branch'))
n_goals = (3,3,3,2)
spec_replacement_dict = {f"{spec}{n}":f"{Spec}" for (spec,Spec),n in zip(spec_map,n_goals)}
# spec_replacement_dict = {f"{spec}{n}":f"{Spec}{n}" for (spec,Spec),n in itertools.product(spec_map,n_goals)}

ablation_map = {'stlpy_uref': 'Prioritize Objective', 'stlpy_safe': 'Prioritize Safety'}

def make_clickable(val):
    # target _blank to open new window
    return '<a target="_blank" href="{}">{}</a>'.format(val, 'Video')

def add_formatted_cols(df):
    def add_stddev(df, new_col, col1, col2, scale=1.0):
        def combine_columns(row):
            return f"{row[col1]*scale:0.2f}"#"\u00B1"f"{row[col2]:2.2f}"  # Example 

        df[new_col] = df.apply(combine_columns, axis=1)
        return df

    # Cols without mean in key
    cols_id = ['finish_rate','mean_path_score']
    cols = ['Finish Rate ↑', 'Mean Score ↑']
    for new_col, col_name in zip(cols, cols_id):
        col1 = col_name
        col2 = f"{col_name}_std"
        scale = 100.0 if col_name == 'finish_rate' else 1.0
        add_stddev(df, new_col, col1, col2, scale=scale)

    # Cols with mean in key
    cols_id = ['safe','plan_time','success']
    cols = ['Safety Rate ↑', 'Planning Time ↓', 'Success Rate ↑']
    for new_col, col_name in zip(cols, cols_id):
        col1 = f"{col_name}_mean"
        col2 = f"{col_name}_std"
        add_stddev(df, new_col, col1, col2)
        
    # Format any remaining cols
    
    df['Planning Time (s) ↓'] = df.apply(lambda row: f"{row['plan_time_mean']:0.2f}", axis=1)
    def format_if_not_nan(val):
        if pd.isna(float(val)):
            return "-"
        else:
            return f"{val:0.2f}"
    df['TtR ↓'] = df.apply(lambda row: format_if_not_nan(float(row['TtR'])), axis=1)
    if ablation_mode:
        df['Planner'] = df.apply(lambda row: ablation_map.get(row['planner'],'ignore'), axis=1)
    else:
        df['Planner'] = df.apply(lambda row: f"{row['planner'].upper()}", axis=1)
    
    df['Spec'] = df['spec'].map(spec_replacement_dict)
    df['Video'] = "https://www.google.com/"
    df.style.format({'Video': make_clickable})
    df['Obs'] = df.apply(lambda row: 'Y' if row['n_obs']>0 else 'N', axis=1)
    df['N'] = df['num_agents']

add_formatted_cols(df)
```

```python
def drop_large(df2drop, df, N=32):
    """Drop all results beyond N"""
    large_agents = df['N'][df['N']>N].unique()
    all_specs=df['Spec'].dropna().unique()
    to_drop = list(itertools.product(all_specs,large_agents))
    return df2drop.drop(to_drop, errors='ignore')

def set_ablation_mode(df):
    if ablation_mode:
        df = df[df['Planner'] != 'ignore']
    else:
        df = df[(df['Planner'] != 'STLPY_SAFE') & (df['Planner'] != 'STLPY_UREF')]
    return df
```

```python
def replace_latex_math(string_to_replace):
    """Simple fn to replace given symbols with the latex alternatives"""
    symbols_to_replace = ['↓', '↑', r'\cline', r'nan']
    latex_code = ['$\downarrow$', r'$\uparrow$' , r'\cmidrule', r'-']
    for symbol, latex in zip(symbols_to_replace, latex_code):
        string_to_replace = string_to_replace.replace(symbol, latex)

    return string_to_replace

# ANY CODE TO FORMAT TABLE HERE
```

```python
import re
# replace html color with latex
def replace_color(text):
    # Define the regex pattern to match `background-color#<hex>`
    pattern = r'\\background-color#([a-fA-F0-9]{6})'

    # Define the replacement function to convert matched hex to uppercase and format it correctly
    def replace_func(match):
        hex_color = match.group(1).upper()
        return f'\\cellcolor[HTML]{{{hex_color}}}'

    # Perform the replacement using re.sub with the replacement function
    new_text = re.sub(pattern, replace_func, text)
    return new_text

def replace_spec(text):
    pattern = r'multirow\[c\].*{(.*?)}'

    # Define the replacement function to format the matched string correctly
    def replace_func(match):
        content = match.group(1)
        print("context", content)
        return f'multirow[t]{{3}}{{*}}[-1em]{{\\STAB{{\\rotatebox[origin=c]{{90}}{{\\footnotesize {content}}}}}}}'

    # Perform the replacement using re.sub with the replacement function
    return  re.sub(pattern, replace_func, text)

def replace_multirow_header(text):
    # Define the regex pattern to match "Planning Time (s) $\downarrow$"
    pattern_planning = r'Planning Time \(s\) \$\\downarrow\$'

    # Define the regex pattern to match "Success Rate $\uparrow$"
    pattern_success = r'Success Rate \$\\uparrow\$'
    
    # Define the regex pattern to match "Success Rate $\uparrow$"
    pattern_ttr = r'TtR \$\\downarrow\$'

    # Define the replacement string for "Planning Time (s) $\downarrow$"
    replacement_planning = r'\\multirowcell{2}{\\shortstack{Planning\\\\Time (s) $\\downarrow$}}'

    # Define the replacement string for "Success Rate $\uparrow$"
    replacement_success = r'\\multirowcell{2}{\\shortstack{Success\\\\Rate $\\uparrow$}}'
    
    replacement_ttr = r'\\multirowcell{2}{\\shortstack{TtR $\\downarrow$}}'

    # Perform the replacement using re.sub
    new_text = re.sub(pattern_planning, replacement_planning, text)
    new_text = re.sub(pattern_ttr, replacement_ttr, new_text)
    return re.sub(pattern_success, replacement_success, new_text)

def add_cmidrule(text):
    pattern = r'(.*\\multirow)'
    
    # Define the replacement string to add \cmidrule{1-21} before the matched line
    replacement = r'\\hline\1'

    # Perform the replacement using re.sub with multiline mode enabled
    return re.sub(pattern, replacement, text, flags=re.MULTILINE)

from functools import reduce
if ablation_mode:
    functions = [replace_color, replace_spec, add_cmidrule, replace_latex_math]
else:
#     functions = [replace_multirow_header,  replace_color, replace_spec, add_cmidrule, replace_latex_math]
    functions = [replace_color, replace_spec, add_cmidrule, replace_latex_math]
    functions = [replace_spec, add_cmidrule, replace_latex_math]

latex_table = styled_final_table.to_latex()
latex_table = final_table.to_latex()
# Apply the functions in a nested manner
result = reduce(lambda v, f: f(v), reversed(functions), latex_table)
```
